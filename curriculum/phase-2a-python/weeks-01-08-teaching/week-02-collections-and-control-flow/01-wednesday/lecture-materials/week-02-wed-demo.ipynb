## Part 1: Lists

In data analysis we constantly need to store many related values together – all order statuses from a marketplace, all product prices in a category, or all customer review scores. Python **lists** give us a flexible container to do exactly that. They keep values in order, allow duplicates, and can hold any type of object.

Let’s dive in with real Olist numbers. Olist has **8** possible order statuses. The distribution is heavily skewed – 97% of orders are delivered, while others like “created” and “approved” are almost negligible. That’s a business insight you can surface with a simple list!

We’ll create a list of statuses and another of their counts, then explore:
- **Indexing** – pick a specific item (first, last). `statuses[0]` gives “delivered”, the dominant outcome.
- **Slicing** – grab a subset (top 3 statuses). `statuses[:3]` returns the three most frequent statuses; you’d show these to your team.
- **Sorting** – `sorted(statuses)` arranges alphabetically, useful for standardised reports.
- **Modifying** – add/remove elements dynamically with `.append()`, `.insert()`, `.remove()`. This mirrors a live workflow: states may be added as you expand to new regions.
- **Aggregations** – use `max`, `min`, `sum`, and `count` on a list of review scores to get key performance indicators.

**Why these operations?** Imagine you’re the ops manager at Jumia. You need a quick report of the most common order outcomes. With lists you can extract that in seconds. Indexing `statuses[0]` tells you the top outcome is “delivered” – great news for customer satisfaction. Slicing `statuses[:3]` gives the three most frequent statuses; you’d present those to your team. Modifying lists (e.g., `append`, `remove`) mirrors how you’d handle a real workflow where states or categories change.

**The review scores example:** `scores = [5, 4, 1, 5, 3, 2, 5, 4, 5, 1]` simulates a small customer feedback batch.
- `max(scores)` = 5 → some customers are thrilled.
- `min(scores)` = 1 → we have unhappy customers, too.
- `sum(scores)/len(scores)` = 3.5 → average rating, a crucial KPI.
- `scores.count(5)` = 4 → four out of ten orders received a perfect score; that’s 40%, a strong start.

In real analysis, you’d drill down further: are 1‑star reviews concentrated in certain product categories? That’s where loops and filtering come in later.

**Business context:** The `counts` list shows 96,478 delivered orders against only 2 “approved”. That suggests Olist’s order processing is nearly instant. Meanwhile, 625 cancelled orders (0.6%) might hide fraud, stock‑outs, or payment failures. Nigerian marketplaces face similar cancellation reasons—tracking these patterns is the first step to reducing churn.

Later, when you work with pandas, you’ll call functions like `.describe()` that do these aggregations automatically. But understanding what’s happening under the hood makes you a stronger, more intuitive analyst.

**Pro tip:** Whenever you see a collection of values, think *list*. Customer IDs? List. Product names? List. Delivery days? List. Get comfortable with the list type now – it’s the foundation of data wrangling.

**Try this mental model:** A list is like a numbered shelf in your warehouse. Shelf position 0 holds a “delivered” box; position -1 holds the last box. You can add new boxes at the end, insert one in the middle, or take one out. Python gives you all those operations with simple, memorable commands.

```python
# Real data from Olist — 8 order statuses and their counts
statuses = ["delivered", "shipped", "canceled", "unavailable",
            "invoiced", "processing", "created", "approved"]
counts   = [96478, 1107, 625, 609, 314, 301, 5, 2]

print(len(statuses))        # 8
print(statuses[0])          # delivered
print(statuses[-1])         # approved
print(statuses[:3])         # ['delivered', 'shipped', 'canceled']
print(sorted(statuses))     # alphabetical order

# Modifying lists
top_states = ["SP", "RJ", "MG"]
top_states.append("RS")
print(top_states)           # ['SP', 'RJ', 'MG', 'RS']
top_states.insert(0, "BA")
print(top_states)           # ['BA', 'SP', 'RJ', 'MG', 'RS']
top_states.remove("BA")
print(top_states)           # ['SP', 'RJ', 'MG', 'RS']

# Useful list operations
scores = [5, 4, 1, 5, 3, 2, 5, 4, 5, 1]
print(max(scores))          # 5
print(min(scores))          # 1
print(sum(scores))          # 35
print(sum(scores)/len(scores))  # 3.5
print(scores.count(5))      # 4
```

## Part 2: For Loops

Manually crunching numbers for 8 statuses would be tiresome—96,478 delivered orders out of 99,441 total… Time to automate! A **`for` loop** repeats a block of code for every item in a collection, turning repetitive tasks into a breeze.

We’ll use `zip()` to create pairs of `(status, count)`. Then we compute each status’s percentage of total orders and display it with a clean f‑string. The output instantly reveals the health of the marketplace: 97.0% delivered is excellent, but 0.6% cancelled and 0.6% unavailable hint at friction points you’d want to investigate—maybe logistics issues in certain states.

**Note on totals:** The five statuses we loop over (delivered, shipped, canceled, unavailable, invoiced) sum to 99,133 orders. That’s 99.6% of the full dataset. The remaining three tiny statuses (0.4%) are omitted here for brevity, but you’ll often see analysts do the same in dashboards—focus on the big slices.

**Inside the loop:**
- `pct = count / total * 100` calculates percentage.
- `f"{status}: {count:,} orders ({pct:.1f}%)"` uses a comma as a thousands separator (`:,`) and rounds to one decimal.
These small formatting touches make your output readable to managers—imagine presenting raw numbers without commas!

**More loop patterns:**
- `range()` generates a sequence of numbers; perfect for score buckets (1–5) or days of the week.
- `enumerate()` gives you an automatic counter alongside each value, ideal for ranking (e.g., top cities).

**Everyday data work:** A loop with `zip()` is the backbone of many summary reports. If you were tasked with breaking down revenue by payment method, you’d loop identically. If Lagos deliveries had a 2% failure rate in certain neighbourhoods, a `for` loop over neighbourhoods would pinpoint the problem in seconds.

**Nigerian angle:** Imagine you run a last‑mile delivery company in Abuja. You have a list of regions and their order volumes. Looping with `zip()` gives you a quick breakdown. If you suspect delays in certain zones, you’d loop over delivery times and filter—next week’s topic (conditionals) will make that even more powerful.

**Beware of off‑by‑one with `range()`:** `range(1,6)` yields 1,2,3,4,5 – the stop value is exclusive. Always test with a small example before applying to a dataset. A classic mistake leads to missing the last element.

**Why the `for` loop is your best friend:** In data analysis, you rarely process a single value; you process hundreds, thousands, or millions. The `for` loop scales gracefully. Write the logic once, and Python runs it for every item. That’s the difference between a two‑hour manual Excel session and a 0.1‑second Python script.

**Try this mental exercise:** Before running the cell, look at the code and predict each line of output. Can you see how `zip()` pairs the first list with the second? How `enumerate()` adds a position number? Visualising data flow inside a loop will help you debug faster later.

```python
# Loop through statuses with counts
statuses = ["delivered", "shipped", "canceled", "unavailable", "invoiced"]
counts   = [96478, 1107, 625, 609, 314]
total = sum(counts)

for status, count in zip(statuses, counts):
    pct = count / total * 100
    print(f"{status}: {count:,} orders ({pct:.1f}%)")

# Expected output:
# delivered: 96,478 orders (97.0%)
# shipped: 1,107 orders (1.1%)
# canceled: 625 orders (0.6%)
# unavailable: 609 orders (0.6%)
# invoiced: 314 orders (0.3%)

# range() loops
print("\nOrder scores 1-5:")
for score in range(1, 6):
    print(f"  Score {score}")

# Loop with index
cities = ["sao paulo", "rio de janeiro", "belo horizonte"]
for i, city in enumerate(cities):
    print(f"{i+1}. {city.title()}")
# 1. Sao Paulo
# 2. Rio De Janeiro
# 3. Belo Horizonte
```

## Part 3: List Comprehensions

Python offers a super‑readable shortcut for building new lists: **list comprehensions**. They combine looping and filtering into a single line, and they often run faster than traditional loops. Mastering them will save you countless lines when cleaning data.

We’ll transform review scores in two ways:
1. **Filter** – keep only positive scores (≥4). The comprehension `[s for s in scores if s >= 4]` replaces a 4‑line loop.
2. **Label** – map each score to a sentiment tag ("Positive", "Neutral", "Negative") using a conditional expression inside the comprehension.

**Side‑by‑side comparison:** The first example shows a standard 4‑line loop that builds a `positives` list. The list comprehension `[s for s in scores if s >= 4]` does the same in one readable line. For small lists the difference is cosmetic, but when you’re processing 50,000 orders the conciseness and performance matter. Comprehensions run at C speed inside Python, making them faster than appending in a loop.

The second transformation uses a conditional expression: `"Positive" if s >= 4 else "Neutral" if s == 3 else "Negative"`. This reads almost like English: “Positive if score >= 4, else Neutral if score == 3, else Negative.” Wrap it in a comprehension and you label the whole list in one pass. The final `labels.count("Positive")` verifies we have 6 positives out of 10.

**Real‑world scenarios:**
- **Data cleaning:** Convert all customer names to title case: `[name.title() for name in names]`
- **Filtering:** Keep orders over $100: `[order for order in orders if order.total > 100]`
- **Recoding:** Map product categories to broader groups: `["Health" if cat in ["beauty_health", "diapers"] else "General" for cat in categories]`
- **Nigerian use case:** Extract all complaints containing “late delivery” from a list of review texts: `[text for text in reviews if "late" in text.lower()]`

These skills scale directly to text analysis and dashboards. In an e‑commerce setting, you might use a comprehension to flag high‑value orders (`order for order in orders if order["price"] > 500`) or to standardise state abbreviations.

**Important:** List comprehensions create a whole new list. For very large data you might eventually prefer generator expressions to save memory, but for now, comprehensions are your go‑to transformation tool.

**Pro tip:** When a comprehension becomes too nested (e.g., multiple `if` conditions or a complex expression inside), break it into a normal loop – readability always beats cleverness. You’ll see this trade‑off throughout your Python journey.

**Try this mental translation:** Whenever you write a traditional loop that appends to a list, ask yourself: “Can this be expressed as a comprehension?” Practice converting back and forth. Soon you’ll think in comprehensions and code with the same fluency a native speaker thinks in their language.

```python
# Transform a list in one line
scores = [5, 4, 1, 5, 3, 2, 5, 4, 5, 1]

# Standard loop version
positives = []
for s in scores:
    if s >= 4:
        positives.append(s)
print(positives)   # [5, 4, 5, 5, 4, 5]

# List comprehension version (same result)
positives = [s for s in scores if s >= 4]
print(positives)   # [5, 4, 5, 5, 4, 5]

# Transform values
labels = ["Positive" if s >= 4 else "Neutral" if s == 3 else "Negative"
          for s in scores]
print(labels)
# ['Positive', 'Positive', 'Negative', 'Positive', 'Neutral',
#  'Negative', 'Positive', 'Positive', 'Positive', 'Negative']

# Verified count
print(labels.count("Positive"))   # 6
```
