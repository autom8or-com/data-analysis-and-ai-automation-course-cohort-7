Think of a list like a row of lockers in a hallway. Each locker has a number (index) starting from 0, and you can put anything inside — numbers, strings, even other lists. In Python, a list is created with square brackets `[]`, items separated by commas. Lists are ordered, meaning the items stay in the order you put them. They’re mutable: you can add, remove, or change items. You can also access items by their position using `list_name[index]`. Negative indices count from the end. Slicing `list[start:stop]` gives you a sub-list. And functions like `len()`, `max()`, `min()`, `sum()` work on lists of numbers.

````
# List of order statuses from Olist (verified data)

statuses = ["delivered", "shipped", "canceled", "unavailable",
            "invoiced", "processing", "created", "approved"]
counts   = [96478, 1107, 625, 609, 314, 301, 5, 2]

# Basic list operations
print("Number of status types:", len(statuses))

print("First status:", statuses[0])

print("Last status:", statuses[-1])

print("First three:", statuses[:3])

print("Alphabetical order:", sorted(statuses))

# Numerical operations on counts
print("Total orders:", sum(counts))

print("Highest count:", max(counts))

print("Lowest count:", min(counts))

print("Average count per status:", sum(counts) / len(counts))

# Alternative – using the scores list from reviews
scores = [5, 4, 1, 5, 3, 2, 5, 4, 5, 1]
print("Review scores:", scores)

print("How many 5-star ratings?", scores.count(5))

print("Average score:", sum(scores) / len(scores))
````

❌ **Common Mistake: Index Out of Range**  
90% of beginners get this error. If you try to access an index that doesn’t exist, Python throws an `IndexError`.

```python
# Wrong way – list has 5 items, index 5 doesn't exist (max index is 4)
print(top_states[5])   # IndexError: list index out of range
```

✅ **The fix:** Always check the length before accessing, or use a try-except block.

```python
# Safe way
index = 5
if index < len(top_states):
    print(top_states[index])
else:
    print("Index out of bounds! List only has", len(top_states), "items.")
```

Now try it: what happens if you use a negative index that’s too large, like `top_states[-6]`? Same error if the negative index is beyond the first item.

Excellent! You now know how to store and fiddle with collections. But what if we need to do something for every item in the list — like calculate the percentage of each order status? We need to revisit our friend, the **for loop**. Let’s dive in.

Imagine you’re a customer service manager at Olist. Your boss asks, “What percentage of orders are in each status?” You could calculate one at a time… but there are 8 statuses! A for loop lets you repeat a task for every item in a list automatically. That’s what we’ll master now.

A for loop in Python works like this: `for item in collection:` — the variable `item` takes on each value in the list (or any iterable) one by one, and the indented block runs each time. You can loop through a single list, or you can loop through two lists simultaneously using `zip()`. The `range()` function is handy for numeric loops, and `enumerate()` gives you both index and item. We’ll use these to analyze our Olist data like a boss.

````
# Order statuses and their counts (verified Olist data)

statuses = ["delivered", "shipped", "canceled", "unavailable", "invoiced"]
counts   = [96478, 1107, 625, 609, 314]
total_orders = sum(counts)


print("=== Olist Order Status Breakdown ===")

for status, count in zip(statuses, counts):
    percentage = (count / total_orders) * 100
    print(f"{status:>12}: {count:>6,} orders  ({percentage:5.1
f}%)")
````

Look how `zip()` paired each status with its count. The formatted output uses `>` for right alignment and `:,` for thousand separators. The percentages show that 97.0% are delivered — no wonder Olist is successful! Notice that `total_orders` was pre-computed; doing it inside the loop would be less efficient. A for loop can be as simple or as powerful as you need. Did you see that `unavailable` and `invoiced` are neck-and-neck? Interesting…

Now we can repeat actions effortlessly. But sometimes we want to transform a whole list in one clean line — that’s where list comprehensions shine. They’re fast, Pythonic, and loved by data analysts. Let’s unlock them!

When your manager asks, 'Give me all review scores that are positive (4 or 5),' you could write a loop, but a list comprehension does it in one line. It’s like a superpower — concise, readable, and runs faster. In data analysis, we often filter and transform lists; list comprehensions are your Swiss Army knife.

A list comprehension is a compact syntax for creating a new list from an existing iterable, with an optional condition. Structure: `[expression for item in iterable if condition]`. You can think of it as a loop that builds a list inline. For example, `[x*2 for x in numbers]` doubles each number. You can even do if-else inside the expression for more complex logic.

````
# Review scores from Olist (verified)

scores = [5, 4, 1, 5, 3, 2, 5, 4, 5, 1]

# Old way: loop
positives_loop = []
for s in scores:
    if s >= 4:
        positives_loop.append(s)

print("Positive scores (loop):", positives_loop)

# One-liner list comprehension
positives_compr = [s for s in scores if s >= 4]
print("Positive scores (comprehension):", positives_compr)

# Verify they're the same
print("Are they equal?", positives_loop == positives_compr)
````

Both give `[5, 4, 5, 5, 4, 5]`. The comprehension is not only shorter but often faster — Python optimizes it internally. Note the structure: `[s for s in scores if s >= 4]`. The first `s` is the expression (just the value itself), then `for`, the iterable, and an `if`. You can read it as: 'give me s for each s in scores if s is >= 4'. Did you notice we used the same variable name `s`? That’s fine; it’s just a placeholder.

❌ **Common Mistake: Forgetting the `[]` when writing a comprehension**  
Beginners often write the comprehension without the outer brackets, then wonder why it’s a generator object or just the `for` statement.  
```python
# Wrong – missing brackets, syntax error or a generator expression
positives = s for s in scores if s >= 4   # SyntaxError
```
✅ **Fix:** Always wrap the comprehension in square brackets for a list. If you need a generator (for memory efficiency), use parentheses `()`, but that’s a later topic.

🎯 **Group Exercise: Analyzing Olist Payment Methods**  
We have two lists from the Olist database:  
`payment_types = ['credit_card', 'boleto', 'voucher', 'debit_card', 'not_defined']`  
`payment_counts = [76795, 19784, 5775, 1529, 3]`  
Your tasks:  
**1.** Calculate and print the total number of payments (expected: 103,886).  
**2.** Using a `for` loop with `zip()`, print each payment type and its percentage of the total (one decimal place). Expected: `credit_card: 73.9%`, etc.

````
payment_types = ['credit_card', 'boleto', 'voucher', 'debit_card', 'not_defined']
payment_counts = [76795, 19784, 5775, 1529, 3]

# Task 1
total_payments = sum(payment_counts)

print("Total payments:", total_payments)

# Task 2
print("\nPayment methods breakdown:")

for p_type, count in zip(payment_types, payment_counts):
    pct = (count / total_payments) * 100
    print(f"{p_type:>13}: {pct:5.1
f}%")
````

**3.** Using a **list comprehension**, create a new list `popular_types` containing only those payment types that have more than 1000 orders. Expected: `['credit_card', 'boleto', 'voucher', 'debit_card']`.  
**4.** Using another list comprehension, create a list `upper_types` that contains all payment type names in **UPPER CASE**.

````
# Task 3
popular_types = [pt for pt, cnt in zip(payment_types, payment_counts)
 if cnt > 1000]
print("Popular types (count > 1000):", popular_types)

# Task 4
upper_types = [pt.upper()
 for pt in payment_types]
print("Uppercase types:", upper_types)
````

**5.** Calculate and print:  
- The average payment count per type. (Expected: 20,777.2)  
- The payment type with the **maximum** count. Use the `max()` function and perhaps `zip()` or index logic.

````
# Task 5
average_count = total_payments / len(payment_types)

print("Average count per type:", average_count)

# Find the type with max count
max_count = max(payment_counts)

max_type = payment_types[payment_counts.index(max_count)]
print(f"Most used: {max_type} with {max_count:,} payments")
````
