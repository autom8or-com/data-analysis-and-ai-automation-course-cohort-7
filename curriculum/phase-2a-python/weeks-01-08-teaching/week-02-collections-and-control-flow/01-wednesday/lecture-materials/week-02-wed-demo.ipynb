## 🧺 What is a list?

A list is like a basket where you can put many items—numbers, strings, even other baskets! In data analysis, you'll use lists all the time:
- A list of product categories
- A list of order prices
- A list of customer states

You can **access** any item by its position (starting from 0), **add** new items, **remove** some, and **sort** the whole basket.

In Nigeria, think of a list as a queue of customers at a Jumia pickup centre. The first person (position 0) gets served first. The last person can be accessed by counting backwards.

Now let's create our first list with real Olist data.

````
# Real data from Olist — the 8 possible order statuses and how many times they appear
statuses = ["delivered", "shipped", "canceled", "unavailable",
            "invoiced", "processing", "created", "approved"]
counts   = [96478, 1107, 625, 609, 314, 301, 5, 2]

# How many different statuses?
print("Number of statuses:", len(statuses))

# Access by position: first and last
print("First status:", statuses[0])
print("Last status (using index -1):", statuses[-1])

# Slicing: get the first three statuses
print("First three:", statuses[:3])

# Sorting alphabetically
print("Sorted alphabetically:", sorted(statuses))
````

### 🔍 What do we learn from this output?

- There are **8** possible order statuses. That's a reasonable number for an e‑commerce platform.
- `statuses[0]` gives `'delivered'` because list indexing starts at **0**, not 1.
- `statuses[-1]` is `'approved'` — the last item. Negative indexing is a handy shortcut.
- Slicing `statuses[:3]` returns a new list with the first three elements. It's like saying "give me everything from the start up to (but not including) position 3".
- `sorted(statuses)` returns a **new** alphabetically ordered list; the original list stays unchanged.

📌 **Did you notice?** The total order count (sum of `counts`) is **99,441**. We'll verify that later.

Now let's learn how to modify a list.

✅ **Checkpoint:** If you have `statuses = ['delivered', 'shipped', 'canceled' ...]`, which index returns `'canceled'`? And how would you get the last two statuses in one go? (Don't run the code – think first!)

````
# Modifying lists in place: working with top‑selling states
# Olist's top 3 states by order volume

top_states = ["SP", "RJ", "MG"]
print("Original:", top_states)

# Append adds one item to the end
top_states.append("RS")
print("After append:", top_states)

# Insert at a specific position (0 = beginning)
top_states.insert(0, "BA")
print("After insert at 0:", top_states)

# Remove a specific value (first occurrence)
top_states.remove("BA")
print("After remove 'BA':", top_states)

# Tuples: immutable cousin of a list
state_tuple = ("SP", "RJ", "MG")
print("\nTuple:", state_tuple)
print("First element of tuple:", state_tuple[0])

# Uncomment the next line to see the error:
# state_tuple[0] = "BA"   # TypeError: 'tuple' object does not support item assignment
````

### ⚠️ Common mistake: trying to modify a tuple

90% of beginners mix up **lists** (mutable) and **tuples** (immutable).

❌ Wrong:
```python
top_states_tuple = ("SP", "RJ", "MG")
top_states_tuple[0] = "BA"   # crashes with TypeError
```

✅ Fix: convert to a list first, modify, then convert back if needed.
```python
top_states_list = list(top_states_tuple)
top_states_list[0] = "BA"
```

Use a tuple when the data should **never change** (e.g., constant categories). Use a list when you plan to add/remove items.

## 🔄 Why do data analysts love loops?

You're handed the `statuses` and `counts` lists. Your manager wants a neat summary:

```
delivered: 96,478 orders (97.2%)
shipped: 1,107 orders (1.1%)
...
```

You could write 8 print statements—but that's brittle. If a new status appears tomorrow, you'd have to edit the code. With a **for loop**, you write the logic once and it handles any number of items.

Think of a for loop like a conveyor belt at a DHL sorting centre in Lagos. Each box (item) passes by, and a worker does the same action on each one. In our code, the worker is the block of code inside the loop.

````
# Using a for loop with zip to process two lists together
statuses = ["delivered", "shipped", "canceled", "unavailable",
            "invoiced", "processing", "created", "approved"]
counts   = [96478, 1107, 625, 609, 314, 301, 5, 2]

total = sum(counts)   # 99,441
print(f"Total orders: {total:,}\n")

# zip pairs elements from statuses and counts one by one
for status, count in zip(statuses, counts):
    pct = count / total * 100
    print(f"{status}: {count:,} orders ({pct:.1f}%)")
````

### 🔍 Output deep dive

- `delivered` dominates: **96,478 orders, 97.0%** of all. That's typical for a mature dataset — most orders eventually get delivered.
- `shipped` and `canceled` are next, but each under 2%.
- The last two statuses, `created` and `approved`, have only a handful of orders. Olist probably captures these very early stages.

**🎯 Surprising?** Even though we used 8 statuses, the loop didn't change at all compared to the 5‑status version. That's the power of Python's `zip` and `for` — they adapt gracefully.

Now, what if we also need the position number of each status (1st, 2nd…)? That's `enumerate`.

✅ **Checkpoint:** What does `zip(statuses, counts)` actually return? Why can't we just do `for status in statuses, counts`?

````
# enumerate() gives us an automatic counter while looping
cities = ["sao paulo", "rio de janeiro", "belo horizonte"]

# Loop with index, starting human‑friendly numbering at 1
for i, city in enumerate(cities):
    print(f"{i+1}. {city.title()}")

print("\n--- Range loop: print scores 1 to 5 ---")
# range(1, 6) goes from 1 to 5 (stop value is exclusive)
for score in range(1, 6):
    print(f"  Score {score}")
````

### ⚠️ Common off‑by‑one mistakes in loops

❌ Forgetting that `range(start, stop)` stops **before** `stop`:
```python
for i in range(1, 5):
    print(i)   # prints 1,2,3,4 — not 5!
```

✅ To include 5, use `range(1, 6)`.

❌ Iterating over two lists without `zip`:
```python
# Wrong! Will iterate over statuses, then counts separately
for status in statuses, counts:
    print(status)  # prints the whole tuple once
```

✅ Use `zip` to pair items side‑by‑side.

## ⚡ List comprehensions: the data analyst's secret weapon

When cleaning data, you often find yourself saying:
> "Give me all scores that are 4 or above"
> or "Convert these city names to title case"

A list comprehension is just a **compact way to build a new list** by applying an expression to each item in an old list, possibly filtering some out.

Think of it like this: you have a basket of oranges (our old list). You want a basket of orange juice bottles (the new list) but only from ripe oranges (filtering). The comprehension is the juicer machine.

Syntax: `[expression for item in old_list if condition]`

````
# Review scores again — we want only positive scores (4 or 5)
scores = [5, 4, 1, 5, 3, 2, 5, 4, 5, 1]

# Old‑school loop way
positives_loop = []
for s in scores:
    if s >= 4:
        positives_loop.append(s)
print("Loop version:", positives_loop)

# List comprehension way — much cleaner!
positives = [s for s in scores if s >= 4]
print("Comprehension:", positives)

# Now transform to labels
labels = ["Positive" if s >= 4 else "Neutral" if s == 3 else "Negative"
          for s in scores]
print("Labels:", labels)
print("Number of Positives:", labels.count("Positive"))
````

### 📊 What does this tell us?

- **6 out of 10 reviews** are positive (4 or 5 stars). Olist customers are generally happy.
- The comprehension gave the exact same result as the loop, but in one line. That's less typing and less chance of a mistake.
- You can even include a **nested if‑else** right inside the expression to map values to labels.

**🔎 Did you notice?** The condition `if s >= 4` comes at the end, after the for‑clause. That's different from a normal if‑statement. The order matters!

Now let's apply this to a real Olist payment analysis.

✅ **Checkpoint:** What would be the output of `[x*2 for x in range(1,4)]`? Think before you peek!

````
# Real Olist payment types
payment_types = ["credit_card", "boleto", "voucher", "debit_card", "not_defined"]
payment_counts = [76795, 19784, 5775, 1529, 3]

# List comprehension: only payment types with count > 1000
popular_types = [pt for pt, pc in zip(payment_types, payment_counts) if pc > 1000]
print("Popular payment types (>1000 orders):", popular_types)

# Another comprehension: all types in UPPERCASE
upper_types = [pt.upper() for pt in payment_types]
print("Uppercase types:", upper_types)
````

### ⚠️ List comprehension pitfalls

❌ Trying to do too much in one comprehension — it becomes unreadable. If the logic is complex, use a for loop.

❌ Forgetting the square brackets — you'll get a generator expression, not a list.

❌ Missing the `if` clause when you need to filter: `[s for s in scores]` gives back the same list (no filter).

✅ Good practice: if your comprehension contains more than one `for` or `if` clause, seriously consider breaking it into a regular loop for clarity.

Now let's try one on your own.

````
# Mini‑challenge: cities that contain a space
cities = ["sao paulo", "rio de janeiro", "belo horizonte", "curitiba"]

# 1. Use a list comprehension to get only cities with a space in their name
cities_with_space = [city for city in cities if ' ' in city]
print("Cities with a space:", cities_with_space)

# 2. Now convert those cities to title case (first letter capital)
title_cities = [city.title() for city in cities_with_space]
print("Title case:", title_cities)
````

````
# Let's start together — load the data and compute totals
payment_types = ["credit_card", "boleto", "voucher", "debit_card", "not_defined"]
payment_counts = [76795, 19784, 5775, 1529, 3]

total_payments = sum(payment_counts)
print(f"Total payments: {total_payments:,}")
# This becomes our denominator for percentages
````

````
# Task 2: Print each type and its percentage of total (using a for loop)
print("Payment breakdown:")
for ptype, pcount in zip(payment_types, payment_counts):
    pct = pcount / total_payments * 100
    print(f"{ptype}: {pcount:,} orders ({pct:.1f}%)")

# Task 3: Using list comprehension, get types with count > 1000
popular = [ptype for ptype, pcount in zip(payment_types, payment_counts) if pcount > 1000]
print("\nTypes with > 1000 orders:", popular)

# Task 4: Uppercase all type names
upper = [ptype.upper() for ptype in payment_types]
print("Uppercase:", upper)
````

- In the list comprehension, mixing up the order of variables inside `zip`. It must be `zip(types, counts)` not the other way around.
