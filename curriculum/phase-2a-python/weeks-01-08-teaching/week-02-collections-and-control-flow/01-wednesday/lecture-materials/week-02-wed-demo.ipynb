## Why This Matters: Making Sense of 99,441 Orders

The Olist dataset contains 99,441 orders spread across 8 different statuses — "delivered" alone accounts for 96,478 orders (97.0% of the total), while "approved" has just 2. There are also 103,886 payment transactions split across 5 payment methods, with credit cards dominating at 73.9%.

Today we build two foundational tools every data analyst needs before touching any library: **collections** (lists and tuples) to store groups of values together, and **iteration** (for loops and list comprehensions) to process them efficiently.

## §3a — Lists: Keeping Multiple Values Under One Name

A list is Python's way of storing multiple items together under a single variable name. Think of it like a column in a spreadsheet: instead of writing `status1 = "delivered"`, `status2 = "shipped"`, `status3 = "canceled"` as eight separate variables — which breaks down immediately when you have 99,441 rows — you write one list and index into it whenever you need a specific item.

Lists are **ordered** (the first item stays first unless you move it) and **mutable** (you can add, remove, or replace items after creation). This makes them the natural tool for storing sequences of category labels, measured values, or any group of items you will process together.

Notice in the Olist data below that `statuses` and `counts` are two parallel lists: position 0 in both refers to "delivered"/96478, position 1 to "shipped"/1107, and so on. Keeping related data in parallel lists is the stepping stone toward the DataFrames you will build in Week 4.

````python
# --- Indexing and slicing ---
print(len(statuses))      # expected: 8
print(statuses[0])        # expected: delivered
print(statuses[-1])       # expected: approved
print(statuses[:3])       # expected: ['delivered', 'shipped', 'canceled']
print(statuses[2:5])      # expected: ['canceled', 'unavailable', 'invoiced']

# Check if a value is in a list
print("canceled" in statuses)   # expected: True
print("refunded" in statuses)   # expected: False

# sorted() returns a new sorted list — the original is unchanged
print(sorted(statuses))   # expected: alphabetical list of all 8 statuses
print(statuses[0])        # expected: delivered  (original still starts with 'delivered')
````

````python
# --- Modifying lists ---
top_states = ["SP", "RJ", "MG"]
top_states.append("RS")
print(top_states)          # expected: ['SP', 'RJ', 'MG', 'RS']

top_states.insert(0, "BA")
print(top_states)          # expected: ['BA', 'SP', 'RJ', 'MG', 'RS']

top_states.remove("BA")
print(top_states)          # expected: ['SP', 'RJ', 'MG', 'RS']

# --- Aggregations on a list of numbers ---
print(max(review_scores))                       # expected: 5
print(min(review_scores))                       # expected: 1
print(sum(review_scores))                       # expected: 35
print(sum(review_scores) / len(review_scores))  # expected: 3.5
print(review_scores.count(5))                   # expected: 4  (five appears 4 times)
````

## §3b — For Loops: Doing Something for Every Item

A for loop says: "perform this action once for every item in this sequence." With 8 order statuses, instead of writing 8 separate print() calls, you write the loop body once and Python repeats it automatically — once per item in the sequence.

Three companion tools make for loops especially powerful in data work. **`zip(list1, list2)`** pairs two lists element-by-element so you can iterate both in one pass — exactly like reading two columns of a spreadsheet side by side. **`range(start, stop)`** generates a sequence of integers on demand without needing a list. **`enumerate(list)`** gives you both the index and the value at each step, so you can number items without managing a counter variable yourself.

Together, `for`, `zip()`, and `enumerate()` cover the vast majority of iteration patterns you will encounter before we introduce pandas. Every time you later call `.iterrows()` or `.apply()` on a DataFrame, you are using the same mental model.

````python
# --- Iterating over two parallel lists with zip() ---
total_orders = sum(counts)
print(f"Total orders: {total_orders:,}")  # expected: Total orders: 99,441

print()
for status, count in zip(statuses, counts):
    pct = count / total_orders * 100
    print(f"{status}: {count:,} orders ({pct:.1f}%)")

# Expected output:
# delivered: 96,478 orders (97.0%)
# shipped: 1,107 orders (1.1%)
# canceled: 625 orders (0.6%)
# unavailable: 609 orders (0.6%)
# invoiced: 314 orders (0.3%)
# processing: 301 orders (0.3%)
# created: 5 orders (0.0%)
# approved: 2 orders (0.0%)
````

````python
# --- range(): generate a sequence of numbers ---
print("Review score scale (1 to 5):")
for score in range(1, 6):
    print(f"  Score {score}")
# expected: Score 1, Score 2, ... Score 5

print()

# --- enumerate(): index + value together ---
# Saves you from writing: i = 0 ... i += 1 manually
cities = ["sao paulo", "rio de janeiro", "belo horizonte"]
print("Top cities by order volume:")
for i, city in enumerate(cities):
    print(f"  {i + 1}. {city.title()}")

# Expected:
#   1. Sao Paulo
#   2. Rio De Janeiro
#   3. Belo Horizonte
````

## §3c — List Comprehensions: The One-Line Loop

A list comprehension is a compact way to create a new list by transforming or filtering an existing one. The pattern reads almost like plain English:

```
[expression   for item in list   if condition]
```

"Give me `[this expression]` for each `[item]` in `[this list]`, but only if `[this condition]` is true." The `if` part is optional — leave it out to transform every item without filtering.

Comprehensions are faster to write and easier to read than the equivalent for-loop-with-append, once you are comfortable with the syntax. They appear constantly in real data work: classifying values, cleaning strings, filtering rows. When you learn pandas in Week 4, you will immediately recognise this pattern inside lambda functions and `.apply()` calls — the same logic, a different wrapper.

````python
# --- Filter: keep only scores 4 or above ---

# Loop version (explicit — shows what the comprehension does under the hood):
positives_loop = []
for s in review_scores:
    if s >= 4:
        positives_loop.append(s)
print(positives_loop)   # expected: [5, 4, 5, 5, 4, 5]

# Comprehension version — identical result, one line:
positives = [s for s in review_scores if s >= 4]
print(positives)        # expected: [5, 4, 5, 5, 4, 5]

# --- Transform: classify each score as a label ---
labels = [
    "Positive" if s >= 4 else "Neutral" if s == 3 else "Negative"
    for s in review_scores
]
print(labels)
# expected: ['Positive', 'Positive', 'Negative', 'Positive', 'Neutral',
#            'Negative', 'Positive', 'Positive', 'Positive', 'Negative']

print(f"Positive: {labels.count('Positive')}")  # expected: Positive: 6
print(f"Neutral:  {labels.count('Neutral')}")   # expected: Neutral:  1
print(f"Negative: {labels.count('Negative')}")  # expected: Negative: 3
````

## §4 — Going Deeper: Tuples — Lists You Cannot Change

A tuple looks like a list but uses parentheses instead of square brackets, and it is **immutable**: once created, you cannot add, remove, or replace its items. That constraint is a feature, not a bug — when you store fixed reference data as a tuple (valid score values, approved status names, geographic coordinate pairs), Python prevents you from accidentally modifying it later in your code. If your code tries, you get an `AttributeError` immediately rather than a silent data corruption hours later.

Tuples also unlock **unpacking**: you can assign each element to a separate variable in one line. This is how functions return multiple values in Python — they return a tuple and the caller unpacks it. You already saw this with `enumerate()`, which yields `(index, value)` tuples on every iteration.

The practical rule: **use a list when the collection will grow or change; use a tuple when the values are fixed and should stay that way.**

````python
# --- Tuples: fixed reference data ---
valid_scores = (1, 2, 3, 4, 5)
print(valid_scores)           # expected: (1, 2, 3, 4, 5)
print(valid_scores[0])        # expected: 1
print(type(valid_scores))     # expected: <class 'tuple'>

# Attempting to modify raises AttributeError — shown as comment:
# valid_scores.append(6)   # AttributeError: 'tuple' object has no attribute 'append'

# --- Tuple unpacking: a function returning multiple values ---
def score_summary(scores):
    return (min(scores), max(scores), round(sum(scores) / len(scores), 1))

low, high, avg = score_summary(review_scores)
print(f"Score range: {low}–{high}, average: {avg}")
# expected: Score range: 1–5, average: 3.5

# --- When to use each ---
top3_states = ("SP", "RJ", "MG")   # fixed reference — won't change
print(top3_states)                  # expected: ('SP', 'RJ', 'MG')

results = []                        # grows as we collect results in a loop
for state in top3_states:
    results.append(state.lower())
print(results)                      # expected: ['sp', 'rj', 'mg']
````

## §5 — Common Mistakes to Avoid

Three mistakes catch almost every beginner when working with lists and loops.

**Mistake 1 — IndexError**: Accessing a position that does not exist. A list with 8 items has valid indices 0–7. Trying `my_list[8]` raises `IndexError: list index out of range`. Negative indexing always works: `my_list[-1]` is always the last item, regardless of length.

**Mistake 2 — `=` vs `==` in conditions**: Python uses `=` for assignment and `==` for comparison. Writing `if s = 5:` is a `SyntaxError` — Python stops immediately before running the line. Always use `==` inside `if` and `elif`.

**Mistake 3 — Reference vs copy**: When you write `b = a` where `a` is a list, `b` does not get its own copy — both names point to the *same* list in memory. Modifying `b` also changes `a`. Use `b = a.copy()` (or `b = a[:]`) when you need an independent copy.

````python
# ── Mistake 1: IndexError ──────────────────────────────────────────────────
short = ["delivered", "shipped", "canceled"]

# WRONG (would crash — shown as comment):
# print(short[5])   # IndexError: list index out of range

# CORRECT — valid indices for a 3-item list are 0, 1, 2 (or -3, -2, -1):
print(short[2])    # expected: canceled
print(short[-1])   # expected: canceled

# ── Mistake 2: = vs == ────────────────────────────────────────────────────────
# WRONG (SyntaxError in Python — shown as comment):
# if short[0] = "delivered":

# CORRECT:
if short[0] == "delivered":
    print("First status confirmed")   # expected: First status confirmed

# ── Mistake 3: reference vs copy ─────────────────────────────────────────────
a = ["SP", "RJ", "MG"]
b = a               # b is the SAME list — not a copy
b.append("RS")
print(a)            # expected: ['SP', 'RJ', 'MG', 'RS']  ← a was changed too!

c = a.copy()        # c is an independent copy
c.append("PR")
print(a)            # expected: ['SP', 'RJ', 'MG', 'RS']  ← a unchanged
print(c)            # expected: ['SP', 'RJ', 'MG', 'RS', 'PR']
````

## §8 — Session Summary

| Concept | Key Syntax | Example |
|---|---|---|
| Create a list | `[item1, item2, ...]` | `statuses = ["delivered", "shipped"]` |
| Index | `list[i]`, `list[-1]` | `statuses[0]` → `"delivered"` |
| Slice | `list[start:stop]` | `statuses[:3]` → first 3 items |
| Modify | `.append()`, `.insert()`, `.remove()` | `top_states.append("RS")` |
| Aggregate | `sum()`, `max()`, `min()`, `.count()` | `sum(counts)` → `99441` |
| Membership | `value in list` | `"canceled" in statuses` → `True` |
| Loop a list | `for item in list:` | `for s in statuses:` |
| Loop two lists | `for a, b in zip(l1, l2):` | `for status, count in zip(...)` |
| Loop with index | `for i, item in enumerate(list):` | `for i, city in enumerate(cities):` |
| Number range | `range(start, stop)` | `range(1, 6)` → 1 2 3 4 5 |
| Filter list | `[x for x in l if cond]` | `[s for s in scores if s >= 4]` |
| Transform list | `[expr for x in l]` | `[s.upper() for s in types]` |
| Immutable sequence | `(item1, item2, ...)` | `valid_scores = (1, 2, 3, 4, 5)` |
| Unpack tuple | `a, b, c = tuple` | `low, high, avg = score_summary(...)` |
