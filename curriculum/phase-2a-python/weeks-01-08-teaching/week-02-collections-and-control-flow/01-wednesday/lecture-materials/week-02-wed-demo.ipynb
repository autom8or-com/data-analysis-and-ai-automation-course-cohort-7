## Part 1: Lists

In data analysis we often need to store many values together – all order statuses from an e‑commerce dataset, for example. Python lists give us a flexible container to do exactly that. Let’s dive in with real Olist numbers.

Olist has **8** possible order statuses, and the distribution is heavily skewed towards delivered orders. We’ll learn how to create lists, access individual elements, slice sections, and modify them on the fly. We’ll also perform common aggregations like `max`, `min`, `sum`, and `count` on a list of review scores – operations you’ll use constantly whether you’re analysing sales in Brazil or Nigeria.

```python
# Real data from Olist — 8 order statuses and their counts
statuses = ["delivered", "shipped", "canceled", "unavailable",
            "invoiced", "processing", "created", "approved"]
counts   = [96478, 1107, 625, 609, 314, 301, 5, 2]

print(len(statuses))        # 8
print(statuses[0])          # delivered
print(statuses[-1])         # approved
print(statuses[:3])         # ['delivered', 'shipped', 'canceled']
print(sorted(statuses))     # alphabetical order

# Modifying lists
top_states = ["SP", "RJ", "MG"]
top_states.append("RS")
print(top_states)           # ['SP', 'RJ', 'MG', 'RS']
top_states.insert(0, "BA")
print(top_states)           # ['BA', 'SP', 'RJ', 'MG', 'RS']
top_states.remove("BA")
print(top_states)           # ['SP', 'RJ', 'MG', 'RS']

# Useful list operations
scores = [5, 4, 1, 5, 3, 2, 5, 4, 5, 1]
print(max(scores))          # 5
print(min(scores))          # 1
print(sum(scores))          # 35
print(sum(scores)/len(scores))  # 3.5
print(scores.count(5))      # 4
```

## Part 2: For Loops

Loops let us automate repetitive tasks. Imagine having to calculate the percentage for each of those 8 order statuses by hand – tedious and error‑prone. A `for` loop does it in a few lines.

We’ll use `zip()` to pair each status with its count, then format the result cleanly with f‑strings. You’ll also see `range()` for numeric sequences and `enumerate()` when you need an automatic counter – perfect for ranking Nigerian states by order volume, for instance. These patterns are the bread and butter of everyday data scripting.

```python
# Loop through statuses with counts
statuses = ["delivered", "shipped", "canceled", "unavailable", "invoiced"]
counts   = [96478, 1107, 625, 609, 314]
total = sum(counts)

for status, count in zip(statuses, counts):
    pct = count / total * 100
    print(f"{status}: {count:,} orders ({pct:.1f}%)")

# Expected output:
# delivered: 96,478 orders (97.0%)
# shipped: 1,107 orders (1.1%)
# canceled: 625 orders (0.6%)
# unavailable: 609 orders (0.6%)
# invoiced: 314 orders (0.3%)

# range() loops
print("\nOrder scores 1-5:")
for score in range(1, 6):
    print(f"  Score {score}")

# Loop with index
cities = ["sao paulo", "rio de janeiro", "belo horizonte"]
for i, city in enumerate(cities):
    print(f"{i+1}. {city.title()}")
# 1. Sao Paulo
# 2. Rio De Janeiro
# 3. Belo Horizonte
```

## Part 3: List Comprehensions

List comprehensions are Python’s elegant shortcut for building a new list from an existing one. They condense a loop and a conditional into a single, readable line. We’ll filter review scores to keep only positives (≥4) and transform them into sentiment labels ("Positive", "Neutral", "Negative"). Mastering this skill will save you many lines of code when you’re cleaning and prepping data for analysis, whether you’re dealing with Brazilian Olist data or Nigerian sales figures.

```python
# Transform a list in one line
scores = [5, 4, 1, 5, 3, 2, 5, 4, 5, 1]

# Standard loop version
positives = []
for s in scores:
    if s >= 4:
        positives.append(s)
print(positives)   # [5, 4, 5, 5, 4, 5]

# List comprehension version (same result)
positives = [s for s in scores if s >= 4]
print(positives)   # [5, 4, 5, 5, 4, 5]

# Transform values
labels = ["Positive" if s >= 4 else "Neutral" if s == 3 else "Negative"
          for s in scores]
print(labels)
# ['Positive', 'Positive', 'Negative', 'Positive', 'Neutral',
#  'Negative', 'Positive', 'Positive', 'Positive', 'Negative']

# Verified count
print(labels.count("Positive"))   # 6
```
