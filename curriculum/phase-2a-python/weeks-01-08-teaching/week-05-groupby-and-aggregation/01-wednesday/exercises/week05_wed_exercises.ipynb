**Exercise 3.** Use `.agg()` on `items` grouped by `seller_id` to compute three metrics in one call: `total_revenue` (sum of price), `avg_price` (mean of price), and `total_items` (count of order_item_id). Then find the top seller by revenue.

```python
# Group items by seller_id and aggregate
seller_stats = items.groupby('seller_id').agg(
    total_revenue=('price', 'sum'),
    avg_price=('price', 'mean'),
    total_items=('order_item_id', 'count')
)

# YOUR CODE HERE: sort by total_revenue descending and show the top seller
# Expected top seller revenue: R$229,472.63
```

**Exercise 4.** Build an `order_summary` table: group `items` by `order_id` and compute `total_price` (sum of price), `total_freight` (sum of freight_value), and `item_count` (count of order_item_id). Then answer: how many orders have more than 1 item, what is the max items in a single order, and the average items per order?

```python
# Build order_summary
order_summary = items.groupby('order_id').agg(
    total_price=('price', 'sum'),
    total_freight=('freight_value', 'sum'),
    item_count=('order_item_id', 'count')
)

# YOUR CODE HERE: orders with >1 item, max item_count, mean item_count
# Expected: orders with >1 item = 9,803; max = 21; avg = 1.14
```
