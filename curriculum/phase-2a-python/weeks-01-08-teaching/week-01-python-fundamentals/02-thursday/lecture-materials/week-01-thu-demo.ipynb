## Safe Conversion with `try`/`except`Real-world data often contains missing values like `'N/A'`. A safe function tries to convert and gives a default (e.g., 0.0) if it fails.This prevents your whole analysis from stopping on a bad value.

```python
# Safe conversion with default
def safe_float(value, default=0.0):
    try:
        return float(value)
    except ValueError:
        return default

print(safe_float("58.90"))   # Expected: 58.9
print(safe_float("N/A"))     # Expected: 0.0

```
