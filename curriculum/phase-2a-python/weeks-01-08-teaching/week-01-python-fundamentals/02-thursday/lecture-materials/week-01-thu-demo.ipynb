## Part 1: More String Methods

We’ll work with strings that mimic real Olist data. The examples below show how to clean, split, and check string content.

```
# Cleaning strings with strip() and replace()

category = "  cama_mesa_banho  "
print(category.strip())                     # cama_mesa_banho
print(category.strip().replace("_", " "))   # cama mesa banho
```

```
# Splitting a timestamp into parts
date_str = "2017-10-02 10:56:33"
parts = date_str.split(" ")

date_only = parts[0]
year = date_only.split("-")[0]
print("Date part:", date_only)

print("Year:", year)
```

```
# Checking string content
status = "delivered"
print("Starts with 'd':", status.startswith("d"))   # True
print("Contains 'cancel':", "cancel" in status)     # False
print("Ends with 'ed':", status.endswith("ed"))     # True

# Identifying cancelled invoices
invoice_normal = "536365"
invoice_cancel = "C536379"
print("Is cancel:", invoice_cancel.startswith("C"))

print("Is cancel:", invoice_normal.startswith("C"))
```

## Part 2: Type Conversion

Data often arrives as strings; we need to convert them for calculations. We'll also create a safe conversion function.

```
# Converting strings to numbers
price_str = "58.90"
price_num = float(price_str)

print("Price + 10:", price_num + 10)   # 68.9

qty_str = "6"
qty_num = int(qty_str)

print("Qty × 3:", qty_num * 3)         # 18
```

```
# Safe conversion with try/except
def safe_float(value, default=0.0):
    try:
        return float(value)

    except ValueError:
        return default

print(safe_float("58.90"))   # 58.9
print(safe_float("N/A"))     # 0.0
```

```
# Converting numbers to strings for output
score = 4
label = "Score: " + str(score)

print(label)                  # Score: 4
```
