## Group Exercise Solutions (Part 3)

### 1. Clean the city name: "sao paulo" → "Sao Paulo"

```
city = "sao paulo"
clean_city = city.title()

print(clean_city)   # Sao Paulo
```

### 2. Clean category name: "cama_mesa_banho" → "Cama Mesa Banho"

```
category = "cama_mesa_banho"
clean_category = category.replace("_", " ").title()

print(clean_category)   # Cama Mesa Banho
```

### 3. Extract date part from timestamp "2017-10-02 10:56:33"

```
timestamp = "2017-10-02 10:56:33"
date_part = timestamp.split(" ")[0]
print(date_part)   # 2017-10-02
```

### 4. Extract the year from "2017-10-02"

```
date_str = "2017-10-02"
year = date_str.split("-")[0]
print(year)   # 2017       (as string)
# Optionally convert to integer:
year_int = int(year)

print(year_int)   # 2017
```

### 5. Convert price "120.65" to float and add 19.99 freight

```
price_str = "120.65"
price_float = float(price_str)

total = price_float + 19.99
print(total)   # 140.64
```

### 6. Check if invoice "C536379" is a cancellation (starts with "C")

```
invoice = "C536379"
is_cancel = invoice.startswith("C")

print(is_cancel)   # True
```

## Weekly Assignment Solutions

### 1. Create variables for a fictional Olist order

```
order_id = "a1b2c3d4e5f67890"
customer_city = "sao paulo"
item_price = 120.65
freight_value = 19.99
review_score = 4
order_status = "delivered"
print("Variables set:", order_id, customer_city, item_price, freight_value, review_score, order_status)
```

### 2. Calculate `total_cost` and `is_expensive`

```
total_cost = item_price + freight_value
is_expensive = total_cost > 200
print("Total cost:", total_cost)       # 140.64
print("Is expensive?", is_expensive)   # False
```

### 3. Apply string methods to `customer_city`

```
# UPPER CASE
print(customer_city.upper())        # SAO PAULO

# Title Case
print(customer_city.title())        # Sao Paulo

# Length of the string
print(len(customer_city))           # 9

# Contains "paulo"?
print("paulo" in customer_city)     # True
```

### 4. `safe_int` function that safely converts a string to int

```
def safe_int(value, default=0):
    try:
        return int(value)
    except ValueError:
        return default

# Tests
print(safe_int("42"))       # 42
print(safe_int("hello"))    # 0
print(safe_int("3.5"))      # 0 (raises ValueError)
```

### 5. Formatted f‑string output

```
output = f"Order {order_id[:8]}... from {customer_city.title()} — Status: {order_status} — Total: R${total_cost:.2f}"
print(output)
# Example: Order a1b2c3d4... from Sao Paulo — Status: delivered — Total: R$140.64
```
