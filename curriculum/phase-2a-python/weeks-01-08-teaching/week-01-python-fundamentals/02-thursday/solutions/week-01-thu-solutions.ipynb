## Exercise 1: City Title Case

"sao paulo" needs to become "Sao Paulo".
Use the `.title()` string method.

```python
city = "sao paulo"
print(city.title())
# Expected: Sao Paulo
```

## Exercise 2: Category Formatting

"cama_mesa_banho" needs to become "Cama Mesa Banho".
Replace underscores with spaces, then apply title case.

```python
category = "cama_mesa_banho"
print(category.replace("_", " ").title())
# Expected: Cama Mesa Banho
```

## Exercise 3: Extract Date

Extract just the date part from "2017-10-02 10:56:33".
Use `.split()` to separate date from time.

```python
timestamp = "2017-10-02 10:56:33"
print(timestamp.split(" ")[0])
# Expected: 2017-10-02
```

## Exercise 4: Extract Year

Extract the year from "2017-10-02".
Split by "-" and get the first part.

```python
date_part = "2017-10-02"
print(date_part.split("-")[0])
# Expected: 2017
```

## Exercise 5: Float Conversion + Math

Convert price_str "120.65" to float, then add 19.99 freight.
Use `float()` to convert the string.

```python
price_str = "120.65"
total = float(price_str) + 19.99
print(total)
# Expected: 140.64
```
