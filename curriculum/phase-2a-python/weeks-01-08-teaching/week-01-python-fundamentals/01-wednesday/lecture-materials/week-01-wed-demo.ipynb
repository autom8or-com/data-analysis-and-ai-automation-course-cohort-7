## Variables: Storing Real Orders
*Business hook:* Imagine you're analyzing thousands of Olist orders. Each order has an ID, price, freight cost, delivery status. Instead of typing these values over and over, you assign them to **variables** – labeled memory boxes that hold data.

Variables make your code readable, reusable, and powerful.

### 📦 What is a Variable?
Think of a variable as a **labeled box**. You put a value inside and stick a label on the box so you can find it later.

Example:
```python
price = 58.90   # A box labeled 'price' holding the number 58.90
```
You can open the box anytime to use the value.

````python
# Simple variable creation and type checkingproduct_id = "X123"price = 58.90in_stock = Trueprint(type(product_id))   # <class 'str'>print(type(price))        # <class 'float'>print(type(in_stock))     # <class 'bool'>
````

### 🔍 Going Deeper: Dynamic Typing & None
Python is **dynamically typed** – you don’t have to declare a type. A variable can even change type later. But be careful: using the wrong type can break your code.

`None` is a special value meaning “no value” or “missing.” It’s common in Olist for cancelled orders.

````python
# Real Olist order dataorder_id = "e481f51cbdc54678b7cc49136f2d6af7"item_price = 58.90freight_value = 13.29order_status = "delivered"is_delivered = Truecancelled_orders = None# Check typesprint(type(order_id))          # <class 'str'>print(type(item_price))        # <class 'float'>print(type(freight_value))     # <class 'float'>print(type(order_status))      # <class 'str'>print(type(is_delivered))      # <class 'bool'>print(type(cancelled_orders))  # <class 'NoneType'>
````

### 🚫 Common Mistake: Type Errors
Trying to **add a string and a number** will crash your program. Python doesn’t know how to combine them.

````python
# This will cause a TypeError!text = "Total: "value = 72.19print(text + value)  # 💥 TypeError: can only concatenate str (not "float") to str
````

````python
# Correct: convert the number to a string firsttext = "Total: "value = 72.19print(text + str(value))        # Total: 72.19# Even better – use f-strings (we'll learn more later)print(f"Total: {value}")       # Total: 72.19
````

## Strings: Working with Text Data
*Business hook:* In the Olist dataset, you’ll work with product descriptions, customer names, and city names like “sao paulo.” Often these need cleaning – capitalizing, replacing parts, extracting substrings.

### 🔤 What is a String?
A string is a sequence of characters – like a necklace of beads, each bead is a letter, digit, or symbol. You can manipulate the necklace: convert to uppercase, slice off a part, replace beads.

Key: strings are **immutable** – you can’t change a bead in place; you create a new string.

````python
city = "sao paulo"# String methodsprint(city.upper())                # SAO PAULOprint(city.title())                # Sao Pauloprint(city.replace("sao", "rio")) # rio pauloprint(len(city))                   # 9print(city.strip())                # 'sao paulo' (no leading/trailing spaces here, but useful)
````

### 🔎 Going Deeper: Indexing & Slicing
You can access individual characters with an index (starting at 0), and extract substrings with slicing. Negative indices count from the end.

````python
order_id = "e481f51cbdc54678b7cc49136f2d6af7"
print(order_id[0])       # e (first character)
print(order_id[-1])      # 7 (last character)
print(order_id[:8])      # e481f51c (first 8 chars)
print(order_id[8:16])    # bdc54678
````

### 🧪 f-Strings: Smart String Formatting
f-strings let you embed variables directly inside strings. Great for creating reports. You can even format numbers with commas, decimals.

````python
state = "SP"
sp_orders = 41746
total_orders = 99441
# Calculate percentage
sp_percent = sp_orders / total_orders * 100
print(f"State {state} has {sp_orders:,} orders, which is {sp_percent:.1f}% of total.")
# Expected: State SP has 41,746 orders, which is 42.0% of total.
````

### 🚫 Common Mistake: Trying to Modify Strings in Place

````python
city = "sao paulo"# city[0] = "S"   # 💥 This will raise TypeError: 'str' object does not support item assignmentprint(city[0])
````

````python
city = "sao paulo"# Correct: create a new string with the desired changecorrect_city = "S" + city[1:]   # "Sao paulo"# Or use .capitalize() or .title()print(city.title())  # Sao Paulo
````
