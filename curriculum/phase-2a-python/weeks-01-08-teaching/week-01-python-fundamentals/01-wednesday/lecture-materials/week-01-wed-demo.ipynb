## What Are Variables?

Think of a variable as a named box that holds a piece of data. Like a shipping label on a package, the name tells you what's inside.

In the Olist dataset, every order has:
- An `order_id` (text) — a unique code
- An `item_price` (number) — how much the product costs
- A `freight_value` (number) — shipping cost
- An `order_status` (text) — e.g., 'delivered'
- A flag (`is_delivered`) — True or False
- Possibly missing data (`None`) for cancelled orders

Let's create these variables with real values from Olist.

```python
# Real Olist order valuesorder_id = 'e481f51cbdc54678b7cc49136f2d6af7'order_status = 'delivered'item_price = 58.90freight_value = 13.29is_delivered = Truecancelled_orders = Noneprint('Order ID:', order_id)print('Status:', order_status)print('Price: R$', item_price)
```

## Checking Data Types

Python automatically assigns a **type** to every value. Types determine what you can do with the data: you can add numbers, but you cannot add a number to a text string. Knowing the type avoids surprises.

Common types:
- `str` → text (order ID, city name)
- `float` → number with decimals (price, freight)
- `bool` → True/False (delivery status)
- `NoneType` → missing/unknown data

The `type()` function tells you the type. Let's check our Olist variables.

```python
print(type(order_id))       # Expected: <class 'str'>print(type(item_price))     # Expected: <class 'float'>print(type(is_delivered))   # Expected: <class 'bool'>print(type(cancelled_orders))  # Expected: <class 'NoneType'>
```

## Doing Math with Olist Data

Python handles arithmetic just like a calculator. In data analysis, you constantly compute totals, percentages, and group items.

Common operators: `+`, `-`, `*`, `/`, `//` (floor division), `%` (remainder).

Let's start by adding an item's price and its freight to get the **total cost**.

```python
total_cost = item_price + freight_valueprint(total_cost)  # Expected: 72.19
```

```python
# Profit margin: (revenue - cost) / revenue * 100revenue = 58.90cost = 40.00margin = (revenue - cost) / revenue * 100print(f'Profit margin: {margin:.1f}%')  # Expected: Profit margin: 32.1%# Division types — useful when counting groupsprint(10 / 3)    # Expected: 3.3333333333333335print(10 // 3)   # Expected: 3   (full groups)print(10 % 3)    # Expected: 1   (leftovers)# Olist example: how many pages of 10 orders from Brazil?total_orders_brazil = 99441pages = total_orders_brazil // 10leftover = total_orders_brazil % 10print(f'{pages} pages and {leftover} orders leftover')  # Expected: 9944 pages and 1 orders leftover
```

## Working with Text (Strings)

Strings are everywhere in Olist: city names, order IDs, product codes. Python gives you hundreds of methods to clean, change, and slice text.

We'll explore:
- Changing case: `.upper()`, `.title()`
- Replacing parts: `.replace()`
- Slicing: getting a part of the string with `[start:end]`
- Formatting: `f-strings` that let you insert variables into text

Let's use a real seller city and order ID.

```python
city = 'sao paulo'
order_id = 'e481f51cbdc54678b7cc49136f2d6af7'

print(city.upper())          # Expected: SAO PAULO
print(city.title())          # Expected: Sao Paulo
print(city.replace('sao', 'rio'))  # Expected: rio paulo
print(len(city))             # Expected: 9
print(city.strip())

# Indexing and slicing
print(order_id[0])           # Expected: e
print(order_id[-1])          # Expected: 7
print(order_id[:8])          # Expected: e481f51c
print(order_id[8:16])        # Expected: bdc54678
```
