### Task 1: Total Order Value

### Task 2: Percentage of Brazil's Orders from SP State

Calculate the percentage by dividing `sp_orders` by `total_orders_brazil`, multiply by 100, and format to two decimal places with an f‑string.

````python
pct = (sp_orders / total_orders_brazil) * 100
print(f"{pct:.2f}%")
````
