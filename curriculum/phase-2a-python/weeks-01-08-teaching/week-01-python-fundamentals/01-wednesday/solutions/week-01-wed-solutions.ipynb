## Given Values from Olist Dataset
These real values were used in the exercise.

```
# Real Olist dataset values
product_id = "4244733e06e7ecb4970a6e2683c13e61"
seller_city = "campinas"
price = 58.90
freight = 13.29
review_score = 4
total_orders_brazil = 99441
sp_orders = 41746
```

### Task 1: Calculate Total Order Value
price + freight

```
# Calculate total order value
total = price + freight
print(total)
# Expected: 72.19
```

### Task 2: Percentage of SP Orders
(sp_orders / total_orders_brazil) * 100

```
# Percentage of Brazil’s total orders from SP state
pct_sp = (sp_orders / total_orders_brazil) * 100
print(f"{pct_sp:.2f}%")
# Expected: 41.98%
```

### Task 3: Title Case for Seller City
seller_city.title()

```
# Print seller_city in title case
city_title = seller_city.title()

print(city_title)
# Expected: Campinas
```

### Task 4: First 8 Characters of product_id
product_id[:8]

```
# Extract the first 8 characters
id_prefix = product_id[:8]
print(id_prefix)
# Expected: 4244733e
```

### Task 5: f-string with Cost Details
Format: "This order costs R$XX.XX including R$XX.XX freight"

```
# Create an f-string for order cost
total = price + freight
print(f"This order costs R${total:.2f} including R${freight:.2f} freight")
# Expected: This order costs R$72.19 including R$13.29 freight
```

### Task 6a: Round total to 2 decimal places
round(total, 2)

```
# Rounding the total
rounded_total = round(total, 2)

print(rounded_total)
# Expected: 72.19
```

### Task 6b: Integer and Float Conversions
int(price), float(freight), float(price)

```
# Type conversions
print(int(price))      # 58
print(float(freight))  # 13.29
print(float(price))    # 58.9
# Expected: 58, 13.29, 58.9
```

### Task 7 (Challenge): format_order Function
Define a function that returns a formatted string: "Order {id} from {city}: R${total} total"

```
def format_order(order_id, city, price, freight):
    total = price + freight
    return f"Order {order_id} from {city.title()}: R${total:.2f} total"

print(format_order(product_id, seller_city, price, freight))
# Expected: Order 4244733e06e7ecb4970a6e2683c13e61 from Campinas: R$72.19 total
```
